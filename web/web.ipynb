```python
from flask import *
from flaskext.mysql import MySQL
import json 

mysql=MySQL()
app  = Flask(__name__)

app.config['MYSQL_DATABASE_USER'] = 'root'
app.config['MYSQL_DATABASE_PASSWORD'] = ''
app.config['MYSQL_DATABASE_DB'] = 'projectdb'
app.config['MYSQL_DATABASE_HOST'] = 'localhost'

mysql.init_app(app)

@app.route('/')
def home():
    cursor=mysql.get_db().cursor()
    cursor.execute("SELECT name FROM projectdb.test")
    names=cursor.fetchall()
    cursor.execute("SELECT * FROM projectdb.data")
    data_table=cursor.fetchall()
    cursor.execute("SELECT count(*) FROM projectdb.data")
    rowCount=cursor.fetchone()[0]
    
    mzmlNames=[]
    for obj in names:
        mzmlNames.append(obj[0])
    
    
    name_target={}
    nameTarget_data={}
    for index in range(rowCount):
        data=data_table[index][3]
        data=data.replace('{"RTs": ','')
        data=data.replace('}','')
        data=data.replace(' "ints": ',',')
        key=str(data_table[index][1])+str(data_table[index][2])
        nameTarget_data[key]=data
        
        if data_table[index][1] not in name_target:  
            name_target[data_table[index][1]]=[str(data_table[index][2])]
        else:
            name_target[data_table[index][1]].append(str(data_table[index][2]))
    
    print(nameTarget_data)
    name_target=json.dumps(name_target)     
    nameTarget_data=json.dumps(nameTarget_data)
    
    
    return render_template('homepage.html',mzmlNames=mzmlNames,data_table=data_table,name_target=name_target,nameTarget_data=nameTarget_data)
    

if __name__ == '__main__':
    app.run()
```
